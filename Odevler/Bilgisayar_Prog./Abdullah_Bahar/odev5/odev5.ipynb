```python
# Dosya işelmleri :
# 'With' sözcüğü kullanılarak yapılabilir -> Bu durumda dosyayı kapatmaya gerek kalmaz (Dosya otomatik kapanır)
# 'With' kullanılmadan da yapılabilir 	  -> Dosya otomatik kapanmaz. (Bu şekilde yaptım)

# Hocam bu size bir nottur : Ödev olarak belirtilen kısımındaki senaryodan bir şey anlaması güç olduğundan işin içine kendi hayal gücümü de kattım.


dosyaAdi_r = "otel.txt" 		# otel verilerinin olunacağı dosya
dosyaAdi_w = "oteller.txt"		# otel vrilerinin yazılacağı dosya

try:
	rDosya = open(dosyaAdi_r, "r")
	wDosya = open(dosyaAdi_w, "w")

	#
	# ekstra bir islem yapmadan dosya for dongusunde satir sait okunur
	#	
	""" 1. Yol
	for satir in rDosya:
		wDosya.write(satir)
	"""
	#
	# dosya okumak icin "read.line" kullanimi.
	# read.line() ile her bir satır teker teker okunur. Her kullanımda otomatik olarak bir sonraki satırı alır.
	# while döngüsü de false olana kadar çalışır. Yani readline()'dan null  döndügünde false olarak algıalnır ve dongu biter
	#
	"""
	satir = rDosya.readline()
	while satir:
		wDosya.write(satir)
		satir = rDosya.readline()
	"""

	#
	# "readlines()" kullanımı
	# bu method, bütün satırları bir liste olarak dondurur
	#
	"""
	satirlar = rDosya.readlines()
	i = 0						# yazdirma yontemi 1 (eski usul dusunce)
	while i < len(satirlar):
		wDosya.write(satirlar[i])
		i = i + 1
	for satir in satirlar:		# yazdiima yontemi 2
		wDosya.write(satir)
	"""
	
	#
	# readlines ve writelines kullanımı
	#
	satirlar = rDosya.readlines()
	if len(satirlar): 			# verilerin olduğu dosyada eleman yoksa buraya girmez
		satirlar[-1] += "\n" 	# metin dosyalarında son satırda genelde \n işareti olamaz. (devamı altta) 
								# Daha sonradan eklencek elemanların bir alt satıra eklenmesi icindir. Yoksa alt satıra değilde yanına yazar.
	print("Eger deger girmke istemiyorsanız bos bir kilde enter diyin") 
	otel = input("Lutfen otelin adını giriniz : ")
	sehir = input("Lutfen otelin konumunu giriniz (Sehir) : ")
	while  otel != "" and sehir != "":
		ekle = otel + "-" + sehir + "\n"		
		if ekle not in satirlar:
			satirlar.append(ekle)
		else:
			print("Girmis oldugunuz degerler zaten mevcut")
		otel = input("Lutfen otelin adını giriniz : ")
		sehir = input("Lutfen otelin konumunu giriniz (Sehir) : ")
	
	satirlar[-1] = satirlar[-1].strip("\n") # son satırda alt satıra geçmesin diye "\n" işaretini kaldırıyorum. 
											# Ayrıca eleman eklenmezse var olan listenin sonuna eklenen "\n" kaldırılır.
	for i in satirlar:
		print(i)

	wDosya.writelines(satirlar) 			# dosyaya burada yazıyoruz
	
	rDosya.close()
	wDosya.close()
except:
	print("Bir hata oluştu")
```

```python
try:
	pop = open("05_worldpop.txt", "r")
	area = open("05_worldarea.txt", "r")
	density = open("world_pop_density.txt", "w")

	popList = pop.readlines()
	areaList = area.readlines()
	densityList = []

	# senaryoya göre aynı sırada aynı ülkelerin olması gereğince satır sayıları eşit varsayılmıştır.
	for i in range(len(popList)):
		# ilgili index'leri bulduk
		nufusIndex = popList[i].find("-")
		alanIndex = areaList[i].find("-")
		# ilgili parametreleri aldık
		nufus = float(popList[i][nufusIndex + 1:]) # "+ 1" ifadesini koymasamda çalışır. Eğer eklemzsem :
		alan = float(areaList[i][alanIndex + 1:])  # "-" ifadesinden itibaren sayıya çevirir. Her ikisi de "-" olduğundan birbirlerine bölümü pozitif oalcaktır
		# hesaplamayı yaptık
		kisiBasiKM = alan / nufus
		# ulke adlarını bulduk
		ulkeAdi = popList[i][0 : nufusIndex]
		# yazdırmak istediğimiz parcaları birleştirdik
		yazdir = ulkeAdi + "-" + str(kisiBasiKM) + " kisi/km^2" + "\n"
		# yazdiracaklarimizi listede topluyoruz
		densityList.append(yazdir)

	# son  elemanın sonunda "\n" olacağında txt'de fazladan bir satır olumuş oluyor.
	# burda ise fazladan satır oluşturmasının önüne geçiliyor
	densityList[-1] = densityList[-1][:len(densityList[-1]) - 1]  # eger liste bos olursa hata alır. Ancak listenin bos olmayacağı varsayılmıştır 
	
	# iste beklenen an. Veriler dosyaya yazdırılıyor
	density.writelines(densityList)
	
	pop.close()
	area.close()
	density.close()
except:
	print("Hata oluştu")
```
